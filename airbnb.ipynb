```python
df["room_type"].unique()
```

```python
z_sum = df.groupby(["room_type"]).agg({'price':'sum'}).reset_index().sort_values(by=['price'])
z_sum
```

From the above results we can able to find that the Entire home/apt type is the highest profit business model

```python
x_max=df.groupby(["room_type"]).agg({'price':'max'}).reset_index().sort_values(by=['price'])
x_max
```

```python
y_min = df.groupby(["room_type"]).agg({"price":min}).reset_index().sort_values(by=['price'])
y_min
```

```python
w = pd.concat([x_max,y_min,z_sum],axis=0,ignore_index=True)
w
```

```python
pplot= pd.concat([x_max,y_min,z_sum], axis=1, ignore_index=True).drop(columns=[2,4,5])
pplot
```

```python
pplot.rename(columns = {0:'Room_type',1: 'max_price', 3:'min_Price'}, inplace = True)
pplot
```
